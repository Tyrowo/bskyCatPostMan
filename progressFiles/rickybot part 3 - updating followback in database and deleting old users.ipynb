# FOLLOWS TABLE FULL UPDATE


so what are we doing here?
When we look at the friends list we have in our db and compare it to our up to date database we're comparing all 5 properties

user_id, user_handle, follow_uri, follow_date, follows_you

* the did and the handle should never be updated.

* the follow_uri should not need to be updated, we could compare it for changes but I don't think that's ever necessary. Usually it will go to None if we stop following them, but in that case the database entry will be deleted. Just in case we'll do a comparison check to what we have in the database with what exists in bluesky, and if there's ever a difference we will see it and I'll have to make another function to handle it.

* the follow_date will not need to be updated ever once it is registered, but new users will be entered with a timestamp of NOW().

* follows_you is the important one to check. we want to make sure that we have up-to-date info on our follows. Either update that they are following you back, or update that they have stopped following you.

from this list we see we need to have 4 queries every time we do a mass update:
1. insert any new users to the db who were not there before
2. update users that were following and now are not
3. update users that were Not following and now are
4. delete any follows that exist in the database but no longer exist in our friends list (either they were deleted or were banned from bluesky)

````python
# set up map - key = did, values = [did, handle, follow_uri, follow_date, follows_you]
db_follows = {}

# !! for some reason when I input my select statements, it's breaking the final column into its own line at the end. So I have to do some extra processing beforehand to clean up the True and False values
split_rows = 'FALSE'.join(select_all_follows.strip().split('\nFALSE'))
split_rows = 'TRUE'.join(split_rows.split('\nTRUE'))
# print(split_rows)

# now we can split the raw output into lines
split_rows = split_rows.split('\n')

for row in split_rows:
  # char to split on is '	' I think it's a tab,
  # DO NOT use the default split(), because we are using a timestamp for the follow_date, which has a space in it. that means we will get an extra column for date+time
  values = row.split('	')
  db_follows[values[0]] = values

# print(db_follows)
# now we have a map of what exists in our database as is
logging_text = logging_text + f'{len(db_follows)} entries were found in the database "follows" table prior to updating.' + LINE_BREAK + LINE_BREAK
````

````python
# next we have to create 3 results lists: two updates, and an insert list. The delete users will then be converted from the remaining map
# also a 4th list to look at any messed up follow links
update_stopped_following = []
update_started_following = []
insert_new_follows = []
follow_link_mismatches = []

# generate our account's follows list
following = client.get_profile(actor=BSKY_USERNAME).follows_count
print(f'currently following {following} users')
next_page = ''
remaining = following
while remaining > 0:
  print (f'starting the next page of follows. {remaining} followers remaining to check.')
  cur_limit = min(remaining, 100)
  remaining -= cur_limit
  follows = client.get_follows(actor=BSKY_USERNAME, cursor= next_page, limit=cur_limit)
  next_page = follows.cursor
  follows = follows.follows
  for follow in follows:
    # need did, user handle, follow_uri, date_added, follows_you
    did = follow.did
    handle = follow.handle
    follow_uri = follow.viewer.following
    date_added = "NOW()"
    # following is whether they are following you, it looks like followed_by is if they are following you back.
    follows_you = 'FALSE' if follow.viewer.followed_by == None else 'TRUE' # need to format it as a string so it matches what we've got

    # now we have our 4 conditions to check for
    if did not in db_follows:
      # need to add to the insert list
      insert_new_follows.append(f"    ('{did}', '{handle}', '{follow_uri}', {date_added}, {follows_you})") # date_added doesn't need quotations but the others
    else:
      # this did exists as a row in our database, so let's do some comparisons
      cur_row = db_follows[did] # cur row is [did, handle, follow_uri, date_added, follows_you] same as above
      if follow_uri != cur_row[2]:
        # I'm not sure what causes this but it did happen a few times.
        print(f'!! follow uris do not match for user {handle}! {follow_uri} != {cur_row[2]}')
        follow_link_mismatches.append([follow_uri, cur_row[2]])
      if follows_you != cur_row[4]:
        # so the follows you doesn't match, which means that if follows_you is true then we need to add this to the stopped following, and otherwise add it to the
        if cur_row[4] == 'TRUE':
          update_stopped_following.append(f"'{did}'")
        else:
          update_started_following.append(f"'{did}'")
      # after the checks we are finished with this entry, so delete it from the map
      del db_follows[did]

    # couldn't use {END_LINE if cur < following else ''} in this print statement for some reason following rounds but the limit gets the exact amount no problem...
logging_text = logging_text + f'Analysis of follows shows: ' + LINE_BREAK
logging_text = logging_text + f'  Started Following: {len(update_started_following)} ' + LINE_BREAK
logging_text = logging_text + f'  Stopped Following: {len(update_stopped_following)} ' + LINE_BREAK
logging_text = logging_text + f'  New Follows Added To DB: {len(insert_new_follows)} ' + LINE_BREAK
logging_text = logging_text + f'  Follow Link Mismatches: {len(follow_link_mismatches)} ' + LINE_BREAK
````

````python
def generate_update_query_no_longer_following(user_list):
  if not user_list:
    print('NO CHANGES NECESSARY')
    return
  users_in = ','.join(user_list)
  print(f'UPDATE {SCHEMA}.{TABLE_FOLLOWS}')
  print(f'SET {COL_FOLLOWSYOU} = FALSE')
  print(f'WHERE {COL_USERID} IN({users_in})')
  print(f'RETURNING {COL_USERHANDLE}, {COL_FOLLOWSYOU}{END_QUERY}')

def generate_update_query_now_following(user_list):
  if not user_list:
    print('NO CHANGES NECESSARY')
    return
  users_in = ','.join(user_list)
  print(f'UPDATE {SCHEMA}.{TABLE_FOLLOWS}')
  print(f'SET {COL_FOLLOWSYOU} = TRUE')
  print(f'WHERE {COL_USERID} IN({users_in})')
  print(f'RETURNING {COL_USERHANDLE}, {COL_FOLLOWSYOU}{END_QUERY}')

def generate_insert_query(rows_list):
  if not rows_list:
    print('NO CHANGES NECESSARY')
    return
  values = ', \n'.join(rows_list)
  print(f'INSERT INTO {SCHEMA}.{TABLE_FOLLOWS} ({COL_USERID}, {COL_USERHANDLE}, {COL_FOLLOWURI}, {COL_FOLLOWDATE}, {COL_FOLLOWSYOU}) ')
  print(f'VALUES ')
  print(values)
  print(f'RETURNING {COL_USERHANDLE}, {COL_FOLLOWDATE}{END_QUERY}')

# for users no longer following or users that no longer exist, weren't found in the comparison
def generate_delete_query(remaining_users_map):
  if not remaining_users_map:
    print('NO CHANGES NECESSARY')
    return
  user_list = [f"'{key}'" for key in remaining_users_map.keys()]
  users_in = ','.join(user_list)
  print(f'DELETE FROM {SCHEMA}.{TABLE_FOLLOWS}')
  print(f'WHERE {COL_USERID} IN({users_in})')
  print(f'RETURNING {COL_USERHANDLE}{END_QUERY}')

def generate_update_follow_uri_query(new_uri, old_uri):
  print(f'UPDATE {SCHEMA}.{TABLE_FOLLOWS}')
  print(f'SET {COL_FOLLOWURI} = {new_uri}')
  print(f'WHERE {COL_FOLLOWURI} = {old_uri}')
  print(f'RETURNING {COL_USERHANDLE}, {COL_FOLLOWURI}{END_QUERY}')
````
