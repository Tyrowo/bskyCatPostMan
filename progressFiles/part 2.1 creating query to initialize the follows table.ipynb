```python
follow_map = {}

following = client.get_profile(actor=BSKY_USERNAME).follows_count
next_page = ''
cur = 0
x = following
# for i in range(100): # so this solved the question - when you reach the last page and use the cursor to go to the next page it just doesn't go anywhere. that's it

# set up query as copy paste lol
# INSERT INTO contacts (first_name, last_name, email)
# VALUES
#     ('Eva', 'Williams', '[[email protected]](../cdn-cgi/l/email-protection.html)'),
#     ('Michael', 'Miller', '[[email protected]](../cdn-cgi/l/email-protection.html)'),
#     ('Sophie', 'Davis', '[[email protected]](../cdn-cgi/l/email-protection.html)')
# RETURNING id;
SCHEMA = 'bsky'
TABLE_FOLLOWS = 'follows'
COL_USERID = 'user_id'
COL_USERHANDLE = 'user_handle'
COL_FOLLOWURI = 'follow_uri'
COL_FOLLOWDATE = 'follow_date'
COL_FOLLOWSYOU = 'follows_you'
END_LINE = ','
END_QUERY = ';'
# print(following)
# print('~~~~~~~~~~~~~~~~~ BEGIN COPYING SQL QUERY HERE~~~~~~~~~~~~~~~~~')
print(f'INSERT INTO "{SCHEMA}"."{TABLE_FOLLOWS}" ({COL_USERID}, {COL_USERHANDLE}, {COL_FOLLOWURI}, {COL_FOLLOWDATE}, {COL_FOLLOWSYOU}) ')
print(f'VALUES ')

while x > 0:
  cur_limit = min(x, 100)
  x -= cur_limit
  follows = client.get_follows(actor='tyrowo.bsky.social', cursor= next_page, limit=cur_limit)
  next_page = follows.cursor
  follows = follows.follows
  for follow in follows:
    cur += 1
    if cur != 1:
      print(END_LINE) # need to finish the previous queries line but don
    # need did, user handle, follow_uri, date_added, follows_you
    did = follow.did
    handle = follow.handle
    follow_uri = follow.viewer.following
    # print(f'| {cur} | did {did} | handle {handle} | following uri {follow_uri} | followed by uri {follows_you} |')
    # from that command we see that the follows are sorted by newest to oldest, so we can fake a date added odering by starting at NOW() (because nothing will be older than now in the db) - interval of x hours where x is their order
    date_added = f"NOW() - interval '{cur} hours'"
    # following is whether they are following you, it looks like followed_by is if they are following you back.
    follows_you = False if follow.viewer.followed_by == None else True


    print(f"    ('{did}', '{handle}', '{follow_uri}', {date_added}, '{follows_you}') ") # date_added doesn't need quotations but the others
    # couldn't use {END_LINE if cur < following else ''} in this print statement for some reason following rounds but the limit gets the exact amount no problem...

print(f'RETURNING {COL_USERHANDLE}{END_QUERY}')

# print('~~~~~~~~~~~~~~~~~ STOP COPYING SQL QUERY HERE~~~~~~~~~~~~~~~~~')
# print(f'there were {cur} follows')
```
